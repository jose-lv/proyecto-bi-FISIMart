#### Celda 2: Configuración del Pipeline Corporativo y Creación de Infraestructura Rígida

```python
import os
import random
from datetime import datetime, timedelta
from faker import Faker

# Ajuste estricto de la semilla analítica corporativa (Semilla 20)
SEMILLA = 20
random.seed(SEMILLA)
np.random.seed(SEMILLA)
fake = Faker('es_ES')
Faker.seed(SEMILLA)

# Parámetros nominales de volumetría requerida por el negocio
N_CLIENTES = 5000
N_PRODUCTOS = 600
N_TIENDAS = 20
N_PROMOCIONES = 40
N_LINEAS_VENTA = 50000

# Creación de la arquitectura de carpetas desde 'notebooks/' usando rutas relativas
os.makedirs("../data/raw", exist_ok=True)
os.makedirs("../data/processed", exist_ok=True)
os.makedirs("../notebooks", exist_ok=True)
os.makedirs("../prompts", exist_ok=True)
os.makedirs("../powerbi", exist_ok=True)
os.makedirs("../informe", exist_ok=True)

print(f"● Pipeline FISIMart S.A.C. inicializado con Semilla {SEMILLA}.")
print("● Topologia de carpetas relativas validada y creada de forma satisfactoria.")
```

#### Celda 3: Construcción de Dimensiones Maestras en Memoria (Cleaned Text)

```python
print("Generando Dimensiones Maestras Libres de Caracteres Especiales...")

# 1. GENERACIÓN: DIM_TIENDA (20 Sucursales garantizando canal omnicanal)
canales_tienda = ["Fisico"] * 18 + ["Online"] * 2
distritos_tiendas = ["Miraflores", "San Isidro", "Santiago de Surco", "San Borja", "La Molina", "Lima Cercado"]
tiendas_pool = []
for i in range(1, N_TIENDAS + 1):
    canal = canales_tienda[i-1]
    nombre_tie = f"FISIMart {canal} - {random.choice(distritos_tiendas) if canal=='Fisico' else 'Ecommerce'}"
    tiendas_pool.append({
        "id_tienda": f"TIE{str(i).zfill(3)}",
        "nombre": nombre_tie,
        "canal": canal,
        "region": "Lima Metropolitana",
        "ciudad": "Lima"
    })
df_tienda_clean = pd.DataFrame(tiendas_pool)

# 2. GENERACIÓN: DIM_PROMOCION (40 Campañas + Registro Técnico de Compra Orgánica)
promociones_pool = [{
    "id_promocion": "PROMO_NONE", 
    "nombre": "Sin Promocion", 
    "tipo": "Ninguno", 
    "descuento_pct": 0.0, 
    "fecha_inicio": datetime(2024, 1, 1), 
    "fecha_fin": datetime(2025, 12, 31)
}]
fecha_base_p = datetime(2024, 1, 1)
for i in range(1, N_PROMOCIONES + 1):
    f_inicio = fecha_base_p + timedelta(days=random.randint(0, 640))
    f_fin = f_inicio + timedelta(days=random.randint(5, 25))
    promociones_pool.append({
        "id_promocion": f"PRM{str(i).zfill(3)}",
        "nombre": f"Campana {random.choice(['Cyber FISI', 'Descuento Especial', 'Aniversario'])} {i}",
        "tipo": random.choice(["Descuento Directo", "Volumen", "Estacional"]),
        "descuento_pct": round(random.uniform(0.05, 0.30), 2),
        "fecha_inicio": f_inicio,
        "fecha_fin": f_fin
    })
df_promocion_clean = pd.DataFrame(promociones_pool)

# 3. GENERACIÓN: DIM_PRODUCTO (600 SKU con Restricción Rígida de Margen)
categorias_map = {
    "Abarrotes": ["Snacks", "Arroz y Menestras", "Aceites", "Conservas"],
    "Tecnologia": ["Accesorios", "Gadgets", "Cargadores"],
    "Hogar": ["Utensilios", "Organizadores", "Decoracion"],
    "Bebidas": ["Gaseosas", "Aguas", "Jugos", "Cervezas"],
    "Cuidado Personal": ["Jabones", "Shampoo", "Dental"]
}
marcas_pool = ["FISIBrand", "Gloria", "InkaCola", "Arroz Superior", "Primor", "Samsung Acc", "Facusa"]
productos_pool = []
for i in range(1, N_PRODUCTOS + 1):
    cat_elegida = random.choice(list(categorias_map.keys()))
    subcat_elegida = random.choice(categorias_map[cat_elegida])
    costo_unitario = round(random.uniform(2.5, 120.0), 2)
    precio_sugerido = round(costo_unitario * random.uniform(1.25, 1.85), 2)  # Cumple precio_lista > costo siempre
    productos_pool.append({
        "id_producto": f"PROD{str(i).zfill(4)}",
        "nombre": f"Producto {subcat_elegida} {i}",
        "categoria": cat_elegida,
        "subcategoria": subcat_elegida,
        "marca": random.choice(marcas_pool),
        "precio_lista": precio_sugerido,
        "costo": costo_unitario
    })
df_producto_clean = pd.DataFrame(productos_pool)

# 4. GENERACIÓN: DIM_CLIENTE (5000 Clientes con vector estructurado de Churn)
distritos_pool = ["Miraflores", "San Borja", "Comas", "Los Olivos", "Santiago de Surco", "San Juan de Lurigancho", "Ate", "Magdalena", "San Miguel"]
clientes_pool = []
for i in range(1, N_CLIENTES + 1):
    sexo = random.choice(["M", "F"])
    edad = random.randint(18, 73)
    f_nacimiento = datetime(2026, 6, 28) - timedelta(days=int(edad * 365.25) + random.randint(0, 350))
    f_alta = datetime(2024, 1, 1) + timedelta(days=random.randint(0, 450))
    
    # Inyección de patrón analítico: Menores ingresos mensuales = Propensión a la deserción
    es_propenso_churn = random.random() < 0.23
    if es_propenso_churn:
        ingresos_estimados = round(random.uniform(1025.0, 2750.0), 2)
        segmento_programa = "Basico"
    else:
        ingresos_estimados = round(random.uniform(2900.0, 12000.0), 2)
        segmento_programa = random.choice(["Platino", "Premium", "Regular"])
        
    clientes_pool.append({
        "id_cliente": f"CLI{str(i).zfill(5)}",
        "nombre": fake.name(),
        "sexo": sexo,
        "fecha_nacimiento": f_nacimiento,
        "edad": edad,
        "distrito": random.choice(distritos_pool),
        "fecha_alta": f_alta,
        "segmento_programa": segmento_programa,
        "ingresos_mensuales_aprox": ingresos_estimados,
        "churn_propensity": es_propenso_churn
    })
df_cliente_clean = pd.DataFrame(clientes_pool)

# Saneamiento de cadenas de Faker para remover acentos latentes en la codificación de nombres
df_cliente_clean["nombre"] = df_cliente_clean["nombre"].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

# 5. GENERACIÓN: DIM_TIEMPO (Análisis analítico temporal de 2 años)
fechas_calendario = [datetime(2024, 1, 1) + timedelta(days=x) for x in range(731)]
feriados_peru = ["01-01", "05-01", "07-28", "07-29", "08-30", "10-08", "11-01", "12-08", "12-25"]
tiempo_pool = []
for f in fechas_calendario:
    tiempo_pool.append({
        "fecha": f,
        "dia": f.day,
        "mes": f.month,
        "trimestre": (f.month - 1) // 3 + 1,
        "anio": f.year,
        "dia_semana": f.strftime("%A"),
        "es_feriado": 1 if f.strftime("%m-%d") in feriados_peru else 0
    })
df_tiempo_clean = pd.DataFrame(tiempo_pool)
df_tiempo_clean["fecha"] = df_tiempo_clean["fecha"].dt.strftime('%Y-%m-%d')

print(f"✓ Catálogos cargados en memoria (Clientes: {len(df_cliente_clean)}, Productos: {len(df_producto_clean)}).")
```

#### Celda 4: Simulación Avanzada de Hechos (Fact_Ventas) con Realismo Estadístico

```python
print("Construyendo matriz transaccional calibrada para Fact_Ventas...")

# A. Distribución Exponencial para Cumplimiento Estricto de Pareto (80/20)
lista_id_productos = df_producto_clean["id_producto"].tolist()
pesos_productos = np.random.exponential(scale=1.0, size=len(df_producto_clean))
pesos_productos[int(len(df_producto_clean) * 0.2):] *= 0.04  # Atenuar la cola larga
pesos_productos /= pesos_productos.sum()

lista_id_clientes = df_cliente_clean["id_cliente"].tolist()
pesos_clientes = np.random.exponential(scale=1.0, size=len(df_cliente_clean))
pesos_clientes /= pesos_clientes.sum()

productos_por_categoria = df_producto_clean.groupby("categoria")["id_producto"].apply(list).to_dict()

lineas_ventas_pool = []
id_boleta_acumulador = 1

# Multiplicadores de estacionalidad mensual del retail peruano
estacionalidad_mensual = {1: 1.3, 2: 0.5, 3: 0.6, 4: 0.9, 5: 1.1, 6: 1.5, 7: 2.3, 8: 0.7, 9: 0.9, 10: 0.8, 11: 1.6, 12: 2.6}

# Convertimos las fechas a objetos datetime reales para asegurar la extracción del mes sin errores
df_tiempo_clean['fecha_dt'] = pd.to_datetime(df_tiempo_clean['fecha'])

print("Generando registros distribuidos día por día...")
# Recorremos los 730 días de forma secuencial
for _, t_row in df_tiempo_clean.iterrows():
    if len(lineas_ventas_pool) >= N_LINEAS_VENTA:
        break
        
    fecha_actual = t_row["fecha_dt"]
    mes_actual = fecha_actual.month  # Extraemos el número de mes (1 al 12) directo del objeto de fecha
    
    # CALIBRACIÓN: Generamos una cantidad pequeña y controlada de boletas por día
    # multiplicada por la estacionalidad del mes para que alcance para los dos años enteros.
    base_tickets = random.randint(12, 18)
    tickets_diarios = int(base_tickets * estacionalidad_mensual[mes_actual])

    for _ in range(tickets_diarios):
        if len(lineas_ventas_pool) >= N_LINEAS_VENTA:
            break
            
        id_cli = np.random.choice(lista_id_clientes, p=pesos_clientes)
        cliente_meta = df_cliente_clean[df_cliente_clean["id_cliente"] == id_cli].iloc[0]
        
        # B. Lógica de Abandono (Churn)
        if cliente_meta["churn_propensity"] and fecha_actual >= datetime(2025, 9, 1):
            continue
            
        id_venta_boleta = f"VTA{str(id_boleta_acumulador).zfill(6)}"
        id_boleta_acumulador += 1
        
        # Carrito de compras acotado (1 a 3 productos por boleta para controlar volumen)
        tamano_canasta = random.choice([1, 2, 2, 3])
        prod_inicial = np.random.choice(lista_id_productos, p=pesos_productos)
        ticket_productos = [prod_inicial]
        
        cat_inicial = df_producto_clean[df_producto_clean["id_producto"] == prod_inicial]["categoria"].iloc[0]
        if cat_inicial == "Abarrotes" and tamano_canasta > 1 and random.random() < 0.78:
            ticket_productos.append(random.choice(productos_por_categoria["Bebidas"]))
            
        while len(ticket_productos) < tamano_canasta:
            prod_secundario = np.random.choice(lista_id_productos, p=pesos_productos)
            if prod_secundario not in ticket_productos:
                ticket_productos.append(prod_secundario)
                
        # Insertar productos en el pool analítico
        for pid in ticket_productos[:tamano_canasta]:
            if len(lineas_ventas_pool) >= N_LINEAS_VENTA:
                break
                
            prod_row = df_producto_clean[df_producto_clean["id_producto"] == pid].iloc[0]
            p_unitario_transaccional = round(prod_row["precio_lista"] * random.uniform(0.95, 1.05), 2)
            unidades_vendidas = random.choice([1, 1, 2])
            
            campanas_activas = df_promocion_clean[
                (df_promocion_clean["fecha_inicio"] <= fecha_actual) & 
                (df_promocion_clean["fecha_fin"] >= fecha_actual) & 
                (df_promocion_clean["id_promocion"] != "PROMO_NONE")
            ]
            
            if not campanas_activas.empty and random.random() < 0.45:
                promo_seleccionada = campanas_activas.sample(1, random_state=SEMILLA).iloc[0]
                id_promocion_fact = promo_seleccionada["id_promocion"]
                descuento_pct_fact = promo_seleccionada["descuento_pct"]
            else:
                id_promocion_fact = "PROMO_NONE"
                descuento_pct_fact = 0.0
                
            lineas_ventas_pool.append({
                "id_venta": id_venta_boleta,
                "fecha": fecha_actual.strftime('%Y-%m-%d'), # Guardar como texto limpio YYYY-MM-DD
                "id_cliente": id_cli,
                "id_producto": pid,
                "id_tienda": f"TIE{str(random.randint(1, N_TIENDAS)).zfill(3)}",
                "id_promocion": id_promocion_fact,
                "cantidad": unidades_vendidas,
                "precio_unitario": p_unitario_transaccional,
                "descuento": descuento_pct_fact
            })

# Limpieza de la columna temporal
df_tiempo_clean.drop(columns=['fecha_dt'], inplace=True, errors='ignore')

df_ventas_clean = pd.DataFrame(lineas_ventas_pool)
df_cliente_clean.drop(columns=["churn_propensity"], inplace=True, errors='ignore')
print(f"✓ Tabla Fact_Ventas estabilizada con éxito: {len(df_ventas_clean)} registros distribuidos homogeneamente.")
```

#### Celda 5: Inyección Controlada de Ruido y Anomalías de Calidad (Entorno Raw)

```python
import numpy as np

print("Contaminando colecciones raw con anomalías e imperfecciones analíticas...")

df_cliente_raw = df_cliente_clean.copy()
df_producto_raw = df_producto_clean.copy()
df_ventas_raw = df_ventas_clean.copy()
df_tienda_raw = df_tienda_clean.copy()
df_tiempo_raw = df_tiempo_clean.copy()
df_promocion_raw = df_promocion_clean.copy()

# CLIENTES: 1.5% de nulos en los ingresos de los clientes
idx_nulos_ingresos = df_cliente_raw.sample(int(len(df_cliente_raw) * 0.015), random_state=SEMILLA).index
df_cliente_raw.loc[idx_nulos_ingresos, "ingresos_mensuales_aprox"] = np.nan

# VENTAS: 1.5% de nulos en el descuento de ventas (omisión de registro)
idx_nulos_descuento = df_ventas_raw.sample(int(len(df_ventas_raw) * 0.015), random_state=SEMILLA).index
df_ventas_raw.loc[idx_nulos_descuento, "descuento"] = np.nan

# PRODUCTOS: 2% de nulos en la columna 'marca' (SKUs genéricos o sin registrar)
idx_nulos_marca = df_producto_raw.sample(int(len(df_producto_raw) * 0.02), random_state=SEMILLA).index
df_producto_raw.loc[idx_nulos_marca, "marca"] = np.nan

# TIENDAS: 1% de nulos en el 'distrito' de las tiendas físicas (error de catastro)
# Filtramos para aplicar solo a tiendas físicas ya que la online usa 'Ecommerce' fijo
idx_tiendas_fisicas = df_tienda_raw[df_tienda_raw["canal"] == "Fisico"].index
idx_nulos_tienda = df_tienda_raw.loc[idx_tiendas_fisicas].sample(int(len(idx_tiendas_fisicas) * 0.01), random_state=SEMILLA).index
df_tienda_raw.loc[idx_nulos_tienda, "nombre"] = np.nan  # Simulamos pérdida del nombre comercial de la sucursal

# PROMOCIONES: 1% de nulos en 'tipo' de promoción
idx_nulos_promo = df_promocion_raw.sample(int(len(df_promocion_raw) * 0.01), random_state=SEMILLA).index
df_promocion_raw.loc[idx_nulos_promo, "tipo"] = np.nan

# TIEMPO: 1% de nulos en la bandera de 'es_feriado' (falla de carga de calendario)
idx_nulos_tiempo = df_tiempo_raw.sample(int(len(df_tiempo_raw) * 0.01), random_state=SEMILLA).index
df_tiempo_raw.loc[idx_nulos_tiempo, "es_feriado"] = np.nan


# Duplicar el 1% de las ventas (errores de envío/doble click en pasarela)
df_duplicados_ventas = df_ventas_raw.sample(int(len(df_ventas_raw) * 0.01), random_state=SEMILLA)
df_ventas_raw = pd.concat([df_ventas_raw, df_duplicados_ventas], ignore_index=True)


# Formato de fechas heterogéneas mixto (2% de afectación)
def mutar_formato_fecha(fecha_val):
    if isinstance(fecha_val, str):
        dt = pd.to_datetime(fecha_val)
    else:
        dt = fecha_val
        
    if random.random() < 0.02:
        variante = random.choice([1, 2])
        return dt.strftime("%d/%m/%Y") if variante == 1 else f"{dt.year}/Dic/{dt.day}"
    return dt.strftime("%Y-%m-%d") if isinstance(fecha_val, str) else dt

df_ventas_raw["fecha"] = df_ventas_raw["fecha"].apply(mutar_formato_fecha)
df_cliente_raw["fecha_alta"] = df_cliente_raw["fecha_alta"].apply(mutar_formato_fecha)

# Inconsistencia tipográfica y errores de tipeo caóticos en Categorías
def contaminar_texto_categorias(val):
    if pd.isna(val): 
        return val
    if random.random() < 0.25:
        if "Bebidas" in val: 
            return random.choice(["Bebidass", "  bebidas ", "BEBIDAS"])
        if "Abarrotes" in val: 
            return random.choice(["  aBaRroTes ", "Abarrotess", "ABARROTES"])
        return f" {val.lower()} "
    return val
df_producto_raw["categoria"] = df_producto_raw["categoria"].apply(contaminar_texto_categorias)

# Outliers severos de operación comercial (0.3% de impacto)
indices_outliers = df_ventas_raw.sample(int(len(df_ventas_raw) * 0.003), random_state=SEMILLA).index
for idx in indices_outliers:
    if random.random() < 0.5:
        df_ventas_raw.at[idx, "cantidad"] = 500  
    else:
        df_ventas_raw.at[idx, "precio_unitario"] = 0.0  

# Ruptura de Integridad Referencial / Registros Huérfanos (0.3% del volumen)
indices_huerfanos = df_ventas_raw.sample(int(len(df_ventas_raw) * 0.003), random_state=SEMILLA).index
df_ventas_raw.loc[indices_huerfanos, "id_producto"] = "PROD9999"

print("✓ Inyección de anomalías de entrada culminada exitosamente (Nulos distribuidos en todas las tablas).")
```

```python
print("=" * 95)
print("             PERFILADO DE DATOS (FISIMart S.A.C. - 2026)             ")
print("=" * 95)

# Mapeo de todas las colecciones en su estado RAW (Crudo)
mapeo_auditoria = {
    "Clientes Crudos (raw)": df_cliente_raw,
    "Productos Crudos (raw)": df_producto_raw,
    "Tiendas Crudas (raw)": df_tienda_raw,
    "Tiempo Crudo (raw)": df_tiempo_raw,
    "Promociones Crudas (raw)": df_promocion_raw,
    "Ventas Crudas (Hechos raw)": df_ventas_raw
}

for titulo, dataframe in mapeo_auditoria.items():
    print(f"\n ENTIDAD: {titulo.upper()}")
    print(f" -> Volumen total de filas encontradas: {len(dataframe)}")
    print(f" -> Cantidad de registros duplicados idénticos: {dataframe.duplicated().sum()} filas")
    
    # 1. CONTAMINANTE: NULOS TOTALES DE LA TABLA
    # Sumamos todos los nulos de todas las columnas y contamos cuántas filas tienen al menos un nulo
    filas_con_nulos = dataframe.isna().any(axis=1).sum()
    print(f" -> Cantidad de filas con valores nulos (NaN): {filas_con_nulos} filas")
    
    # 2. DIAGNÓSTICO DE OTROS CONTAMINANTES (Conteos absolutos)
    print(" -> Diagnóstico de Otras Anomalías detectadas:")
    
    # A. Fechas con formato corrupto
    columnas_fecha = [col for col in dataframe.columns if "fecha" in col]
    total_fechas_corruptas = 0
    for col in columnas_fecha:
        fechas_mal_formato = dataframe[~dataframe[col].astype(str).str.match(r'^\d{4}-\d{2}-\d{2}$') & dataframe[col].notna()][col]
        total_fechas_corruptas += len(fechas_mal_formato)
    
    if total_fechas_corruptas > 0:
        print(f"    [Formato] Filas con fechas fuera del estándar YYYY-MM-DD: {total_fechas_corruptas} registros")
    else:
        if len(columnas_fecha) > 0:
            print("  [Formato] 0 filas con problemas de formato de fecha.")

    # B. Texto Sucio / Variantes en Categorías
    if "categoria" in dataframe.columns:
        categorias_unicas_raw = dataframe["categoria"].dropna().nunique()
        categorias_unicas_limpias = dataframe["categoria"].dropna().astype(str).str.strip().str.lower().nunique()
        variantes_redundantes = categorias_unicas_raw - categorias_unicas_limpias
        if variantes_redundantes > 0:
            print(f"   [Tipografía] Variantes redundantes de texto creadas por errores de tipeo/espacios: {variantes_redundantes} variantes")
        else:
            print("    [Tipografía] 0 problemas de texto detectados en categorías.")

    # C. Outliers de Negocio (Precios cero y Cantidades excesivas)
    if "precio_unitario" in dataframe.columns:
        precios_cero = (dataframe["precio_unitario"] == 0).sum()
        if precios_cero > 0:
            print(f"    [Negocio] Filas con 'precio_unitario' igual a 0.00: {precios_cero} registros")
            
    if "cantidad" in dataframe.columns:
        cantidades_anormales = (dataframe["cantidad"] == 500).sum()
        if cantidades_anormales > 0:
            print(f"    [Inventario] Filas con 'cantidad' atípica (igual a 500): {cantidades_anormales} registros")

    print("-" * 60)

# 3. EVALUACIÓN DE INTEGRIDAD REFERENCIAL
set_productos_maestros = set(df_producto_clean["id_producto"])
lineas_huerfanas_conteo = df_ventas_raw[~df_ventas_raw["id_producto"].isin(set_productos_maestros)]

print("\n" + "=" * 95)
print("🔍 EVALUACIÓN GLOBAL DE INTEGRIDAD REFERENCIAL")
if len(lineas_huerfanas_conteo) > 0:
    print(f" -> ALERTA CRÍTICA: Se encontraron {len(lineas_huerfanas_conteo)} registros huérfanos en 'Fact_Ventas'.")
    print(f"    (Filas que apuntan al producto inexistente 'PROD9999')")
else:
    print(" -> Integridad Referencial 100% sólida entre Hechos y Dimensiones.")
print("=" * 95)
```
